```python
from google.colab.patches import cv2_imshow
import cv2

img = cv2.imread("img3.jpeg")
#cv2_imshow(img)
cv2.waitKey(0)
```

```python
canny=cv2.Canny(img, 125, 125)
cv2_imshow(canny)
cv2.waitKey(0)
```

```python
import numpy as np
blank=np.zeros(img.shape, dtype='uint8')
cv2_imshow(blank)
```

```python
# another way to find contours, instead of canny
gray=cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
ret, thresh = cv2.threshold(gray, 125, 255, cv2.THRESH_BINARY)
cv2_imshow(thresh)
cv2.waitKey(0)
```

```python
contours, hierarchies = cv2.findContours(thresh, cv2.RETR_LIST, cv2.CHAIN_APPROX_NONE)
# contours, hierarchies = cv2.findContours(canny, cv2.RETR_LIST, cv2.CHAIN_APPROX_SIMPLE)
# contours, hierarchies = cv2.findContours(canny, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)

# hierarchies is what the Open CV uses to find the contours.
print(f'{len(contours)} contour(s) found:')
```

```python
# drawing contours on blank images
cv2.drawContours(blank, contours, -1, (0,0,255), 0)
cv2_imshow(blank)
```
