```python
# face detection is done essentially by a classifier, an algo to tell if face is there or not
```

```python
from google.colab.patches import cv2_imshow
import cv2
import numpy as np

img = cv2.imread("rdj2.jpg")
resized_img=cv2.resize(img, (900,700))

cv2_imshow(resized_img)
```

```python
gray=cv2.cvtColor(resized_img, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray)
```

```python
# reads all the 33k lines and stores it in the variable called haar_cascade
haar_cascade = cv2.CascadeClassifier('haar_cascade_default_file.xml')
```

```python
faces_rect = haar_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=3)
```

```python
print(f'Number of faces found={len(faces_rect)}')
```

```python
for (x, y, w, h) in faces_rect:
  cv2.rectangle(resized_img, (x, y), (x+w, y+h), (0, 255, 0), thickness=2)

cv2_imshow(resized_img)
```

### For group

```python
img2 = cv2.imread("grp2.jpg")
resized_img2=cv2.resize(img2, (900,700))

cv2_imshow(resized_img2)
```

```python
gray2=cv2.cvtColor(resized_img2, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray2)

faces_rect2 = haar_cascade.detectMultiScale(gray2, scaleFactor=1.1, minNeighbors=3)

print(f'Number of faces found={len(faces_rect2)}')
for (x, y, w, h) in faces_rect2:
  cv2.rectangle(resized_img2, (x, y), (x+w, y+h), (0, 255, 0), thickness=2)

cv2_imshow(resized_img2)
```

```python
# haar cascade is very sensitive to noise and hence counts either too many faces or less than the actual.
```
