```python
from google.colab.patches import cv2_imshow
import cv2
import numpy as np

blank = np.zeros((400, 400), dtype='uint8')
rectangle = cv2.rectangle(blank.copy(), (30, 30), (370, 370), 255, -1)
circle = cv2.circle(blank.copy(), (200, 200), 200, 255, -1)

cv2_imshow(rectangle)
print(" ")
cv2_imshow(circle)

cv2.waitKey(0)
```

### Bitwise

```python
# AND operator
bitwise_and = cv2.bitwise_and(rectangle, circle)
cv2_imshow(bitwise_and)
```

```python
# OR operator
bitwise_or = cv2.bitwise_or(rectangle, circle)
cv2_imshow(bitwise_or)
```

```python
# XOR operator
bitwise_xor = cv2.bitwise_xor(rectangle, circle)
cv2_imshow(bitwise_xor)
```

```python
# NOT operator
bitwise_not = cv2.bitwise_not(circle)
cv2_imshow(bitwise_not )
```

# *Masking*
###### dimension of mask has to be of same size than that of image


```python
from google.colab.patches import cv2_imshow
import cv2
import numpy as np

img = cv2.imread("rdj2.jpg")
resized_img=cv2.resize(img, (900,700))
cv2_imshow(resized_img)
cv2.waitKey(0)
```

```python
# basic mask implementation
blank=np.zeros(resized_img.shape[:2], dtype='uint8')   # blank is same shape as of image
mask=cv2.circle(blank, (resized_img.shape[1]//2, resized_img.shape[0]//2), 250, 255, -1)
cv2_imshow(mask)
```

```python
# creating masked image
masked = cv2.bitwise_and(resized_img, resized_img, mask = mask)
cv2_imshow(masked)
```

```python
rectangle = cv2.rectangle(blank.copy(), (300, 300), (700, 700), 255, -1)
masked2 = cv2.bitwise_and(resized_img, resized_img, mask = rectangle)
cv2_imshow(masked2)
```
