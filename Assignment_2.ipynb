```python
# Q7
import seaborn as sns
#Iris-setosa is categoical data so convert it to numeric
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder() # --------->  initialize the data
encoder.fit(data['Species'])
data['Species'] = encoder.transform(data['Species']) # --------->  chaning the column and replacing the words in species with numeric data
```

```python
f, ax = plt.subplots(figsize = (7, 5))
corr = data.corr()
# heatmap is present in seaborn 
sns.heatmap(corr, ax=ax, annot=True)

#sns.heatmap(data, cmap='rainbow')
plt.show()
```

The heat map shows us the correlation between the two features and how much the two features are dependent on each other. The more the number of close to 1 higher the two features are correlated and closer to 0 or negative they are not relevent for training the model are they are not correlated and hence they are not taken into the model training.
Heat map are great at feature selection

```python
# Q9

import statsmodels.api as sm

# Create the Q-Q plot
sm.qqplot(data, line='q')

# Set plot title and labels
plt.title('Q-Q Plot')
plt.xlabel('Theoretical Quantiles')
plt.ylabel('Sample Quantiles')

# Display the plot
plt.show()
```

A Q-Q plot compares the quantiles of the dataset with the quantiles of a theoretical distribution. It helps assess if the data follows the assumed distribution and detects departures from linearity or outliers.
